```python
mydat3 = pd.read_csv(io.BytesIO(uploaded['precip.csv']), encoding="shift-jis")
print(mydat3)
```

```python
scores = mydat3['Precip']
print(scores)
bins = np.linspace(3500, 7500, 11)
print(bins)

freq = scores.value_counts(bins=bins, sort=False)
print(freq)
```

```python
class_value = (bins[:-1] + bins[1:]) / 2 #階級値
rel_freq = freq / scores.count() #相対度数
cum_freq = freq.cumsum() #累積度数
rel_cum_freq = rel_freq.cumsum()
```

```python
dist = pd.DataFrame (
    {
        "階級値": class_value,
        "度数": freq,
        "相対度数": rel_freq,
        "累積度数": cum_freq,
        "相対累積度数": rel_cum_freq,
    },
    index=freq.index
)
dist
```
